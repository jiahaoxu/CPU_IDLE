## Hypothesis: column 3 is the resource usage (%)

### create a instMap

**key: instance name**  
**val: {timestamp: resource usage}**

```python
import collections
instMap = collections.defaultdict(dict)
instUsageMap = collections.defaultdict(list)
```

```python
for row in df2.iterrows():
    instMap[row[0]][row[1][2]] = row[1][3]
    instUsageMap[row[0]].append(row[1][3])
```

```python
len(instMap)
```

```python
len(instUsageMap)
```
